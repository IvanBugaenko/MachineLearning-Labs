```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X1 = np.linspace(-2, 0, 100) 
Y1 = (lambda x: x * 0)(X1)

X2 = np.linspace(0, np.pi, 100) 
Y2 = np.cos(X2 / 2) * 0.5

X3 = np.linspace(np.pi, 5, 100) 
Y3 = (lambda x: x * 0 )(X3)
```

```python
plt.plot(X1, Y1, color='r')
plt.plot(X2, Y2, color='r')
plt.plot(X3, Y3, color='r')
plt.axvline(x=0, ymin=0.05, ymax=0.95, ls='--', color='gray')
plt.plot(0, 0, color='r', marker='o')
```
