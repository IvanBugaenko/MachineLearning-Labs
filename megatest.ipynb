```python
import pandas as pd
import numpy as np
```

```python
table = pd.read_csv("pz4.csv", encoding='utf-8')
table.columns = ["Регион", *list(range(2003, 2020, 1))]
table
```

```python
# Разбиение табличек

l = table.shape[0]
I = []

for num, row in enumerate(table["Регион"]):
    if row == "#####":
        I.append(num)
I.append(l + 1)

razb = []
for i in range(len(I) - 1):
    razb.append((I[i] + 1, I[i + 1] - 1))

print(razb)
```

```python
new_cols = []

for i in range(len(razb)):
    temp_cols = []
    time_table = table.loc[razb[i][0]: razb[i][1]]
    time_table = time_table.set_index(time_table["Регион"])
    time_table = time_table.drop(['Регион'], axis=1)
    for row in time_table.index:
        for col in time_table.columns:
            temp_cols.append([row, col, time_table.loc[row][col]])
    
    if i == 0:
        new_cols = np.array(temp_cols)
    else:
        new_cols = np.c_[new_cols, np.array(temp_cols)]

    # print(temp_cols, sep='\n')

print(new_cols.shape, sep='\n')
```

```python
t = pd.DataFrame(new_cols)
t.drop(columns=[3, 4, 6, 7, 9, 10, 12, 13, 15, 16, 18, 19, 21, 22], inplace=True)
```

```python
t.columns = ["Регион", "Год", "Доходы гостиниц и аналогичных средств размещения", 
               "Число гостиниц и средств размещения", 
               "Численность размещенных лиц в гостиницах и ананлогичных средствах размещения",
               "Число ночевок в гостиницах и ананлогичных средствах размещения",
               "Средняя численность работников, включая внешних совместителей",
               "Число туристских фирм", "Средства, поступившие от реализации туристкого продукта",
               "Число реализованных турпакетов населению"]
t
```

```python
t.to_csv("huita.csv", index=False)
```
