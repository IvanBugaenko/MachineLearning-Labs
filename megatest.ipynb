```python
import pandas as pd
```

```python
table = pd.read_csv("pz4.csv", encoding='utf-8')
table.columns = ["Регион", *list(range(2003, 2020, 1))]
table
```

```python
# Разбиение табличек

l = table.shape[0]
I = []

for num, row in enumerate(table["Регион"]):
    if row == "#####":
        I.append(num)
I.append(l + 1)

razb = []
for i in range(len(I) - 1):
    razb.append((I[i] + 1, I[i + 1] - 1))

print(razb)
```

```python
table.loc[razb[0][0]: razb[0][1]]
```
